```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_theme(style="whitegrid")

df = pd.read_csv("C:\\Users\\tekam\\Downloads\\Data Science\\Data_Analyst_Portfolio_Project\\data\\raw\\investor_transactions_clean.csv")

display(df.head())
```

```python
state_df = (

    df.groupby("state")["amount_inr"]

      .sum()

      .sort_values(ascending=True)

      .reset_index()

)

display(state_df.head())
```

```python
plt.figure(figsize=(12,8))

sns.barplot(

    data=state_df,

    y="state",

    x="amount_inr"

)

plt.title(

    "State-wise SIP Amount",

    fontsize=18,

    weight="bold"

)

plt.xlabel("Total SIP Amount (₹)")

plt.ylabel("State")

plt.tight_layout()

plt.show()
```

```python
plt.figure(figsize=(7,7))

tier = df["city_tier"].value_counts()

plt.pie(

    tier,

    labels=tier.index,

    autopct="%1.1f%%",

    startangle=90

)

plt.title(

    "T30 vs B30 Investor Distribution",

    fontsize=18,

    weight="bold"

)

plt.show()
```
