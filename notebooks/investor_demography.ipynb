```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_theme(style="whitegrid")

df = pd.read_csv(r"C:\Users\tekam\Downloads\Data Science\Data_Analyst_Portfolio_Project\data\raw\investor_transactions_clean.csv")

df.head()
```

```python
print(df.info())

print(df.isnull().sum())

print(df.shape)
```

```python
plt.figure(figsize=(8,8))

age_counts = df["age_group"].value_counts()

plt.pie(

    age_counts,

    labels=age_counts.index,

    autopct="%1.1f%%",

    startangle=90

)

plt.title(

    "Investor Age Group Distribution",

    fontsize=18,

    weight="bold"

)

plt.show()
```

```python
plt.figure(figsize=(12,6))

sns.boxplot(

    data=df,

    x="age_group",

    y="amount_inr"

)

plt.title(

    "SIP Amount Distribution by Age Group",

    fontsize=18,

    weight="bold"

)

plt.xlabel("Age Group")

plt.ylabel("Amount (₹)")

plt.xticks(rotation=20)

plt.show()
```

```python
plt.figure(figsize=(7,7))

gender = df["gender"].value_counts()

plt.pie(

    gender,

    labels=gender.index,

    autopct="%1.1f%%",

    startangle=90

)

plt.title(

    "Gender Distribution",

    fontsize=18,

    weight="bold"

)

plt.show()
```

```python
import os

os.makedirs("dashboard", exist_ok=True)

# Age Group Pie
plt.figure(figsize=(8,8))
plt.pie(
    df["age_group"].value_counts(),
    labels=df["age_group"].value_counts().index,
    autopct="%1.1f%%",
    startangle=90
)
plt.title("Investor Age Group Distribution")
plt.savefig("dashboard/age_group_distribution.png", dpi=300, bbox_inches="tight")
plt.close()

# Box Plot
plt.figure(figsize=(12,6))
sns.boxplot(data=df, x="age_group", y="amount_inr")
plt.title("SIP Amount by Age Group")
plt.savefig("dashboard/sip_amount_boxplot.png", dpi=300, bbox_inches="tight")
plt.close()

# Gender Pie
plt.figure(figsize=(7,7))
plt.pie(
    df["gender"].value_counts(),
    labels=df["gender"].value_counts().index,
    autopct="%1.1f%%",
    startangle=90
)
plt.title("Gender Distribution")
plt.savefig("dashboard/gender_distribution.png", dpi=300, bbox_inches="tight")
plt.close()

print("✅ All charts saved in dashboard folder.")
```
