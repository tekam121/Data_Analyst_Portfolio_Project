```python
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go

sip = pd.read_csv(r"C:\Users\tekam\Downloads\Data Science\Data_Analyst_Portfolio_Project\data\raw\04_monthly_sip_inflows.csv")

sip.head()
```

```python
# Convert month to datetime

sip["month"] = pd.to_datetime(sip["month"])

# Sort by month

sip = sip.sort_values("month")

# Check

print(sip.info())

display(sip.head())
```

```python
fig = px.line(

    sip,

    x="month",

    y="sip_inflow_crore",

    markers=True,

    title="Monthly SIP Inflow Trend (Jan 2022 - Dec 2025)"

)

fig.update_traces(

    line=dict(width=4),

    marker=dict(size=7)

)

fig.update_layout(

    template="plotly_white",

    width=1400,

    height=700,

    xaxis_title="Month",

    yaxis_title="SIP Inflow (₹ Crore)",

    hovermode="x unified"

)
```

```python
highest = sip.loc[sip["sip_inflow_crore"].idxmax()]

fig.add_annotation(

   x = highest["month"].strftime("%Y-%m-%d"),

    y=highest["sip_inflow_crore"],

    text=f"₹{highest['sip_inflow_crore']:,} Cr<br>All-Time High",

    showarrow=True,

    arrowhead=3,

    arrowsize=2,

    arrowwidth=2,

    arrowcolor="red",

    bgcolor="yellow",

    bordercolor="red",

    font=dict(

        size=12,

        color="black"

    )

)
```

```python
fig.add_vrect(

    x0="2025-12-01",

    x1="2025-12-31",

    fillcolor="green",

    opacity=0.15,

    line_width=0

)

fig.add_annotation(

    x="2025-12-15",

    y=max(sip["sip_inflow_crore"])*0.92,

    text="Dec 2025",

    showarrow=False,

    font=dict(

        size=14,

        color="green"

    )

)
```

```python
import os

os.makedirs("dashboard", exist_ok=True)

fig.write_html(

    "dashboard/monthly_sip_inflow.html"

)

'''fig.write_image(

    "dashboard/monthly_sip_inflow.png",

    width=1800,

    height=900,

    scale=2

)
'''
fig.write_html("dashboard/monthly_sip_inflow.html")
fig.show()

print("Dashboard Saved Successfully")
```
