```python
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go
import os

print("Libraries Loaded Successfully")
```

```python
folio = pd.read_csv(r"C:\Users\tekam\Downloads\Data Science\Data_Analyst_Portfolio_Project\data\raw\06_industry_folio_count.csv")

folio["month"] = pd.to_datetime(folio["month"])

folio = folio.sort_values("month")

display(folio.head())

print(folio.shape)
```

```python
fig = px.line(

    folio,

    x="month",

    y="total_folios_crore",

    markers=True,

    title="Industry Folio Count Growth (2022–2025)"

)

fig.update_traces(

    line=dict(width=4),

    marker=dict(size=8)

)

fig.update_layout(

    template="plotly_white",

    width=1400,

    height=700,

    hovermode="x unified",

    xaxis_title="Month",

    yaxis_title="Total Folios (Crore)",

    title={

        "text":"<b>Industry Folio Count Growth (2022–2025)</b>",

        "x":0.5

    }

)
```

```python
milestones = [

    (13.26, "Jan 2022"),
    (15.00, "15 Cr"),
    (20.00, "20 Cr"),
    (25.00, "25 Cr"),
    (26.12, "Dec 2025")

]

for value, label in milestones:

    row = folio.iloc[(folio["total_folios_crore"]-value).abs().argsort()[:1]]

    x = row["month"].iloc[0].strftime("%Y-%m-%d")

    y = row["total_folios_crore"].iloc[0]

    fig.add_annotation(

        x=x,

        y=y,

        text=f"{label}<br>{y:.2f} Cr",

        showarrow=True,

        arrowhead=2,

        bgcolor="yellow",

        bordercolor="black",

        font=dict(size=11)

    )
```

```python
fig.update_xaxes(

    rangeslider_visible=True,

    rangeselector=dict(

        buttons=[

            dict(count=1,label="1Y",step="year",stepmode="backward"),

            dict(count=2,label="2Y",step="year",stepmode="backward"),

            dict(label="All",step="all")

        ]

    )

)

fig.show()
```

```python
os.makedirs("dashboard", exist_ok=True)

fig.write_html(

    "dashboard/industry_folio_growth.html"

)

print("✅ HTML Dashboard Saved Successfully")

# PNG Export (Optional)

# fig.write_image(
#     "dashboard/industry_folio_growth.png",
#     width=1800,
#     height=900,
#     scale=2
# )
```
